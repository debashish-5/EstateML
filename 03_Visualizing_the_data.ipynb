```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('housing.csv')
```

```python
df

# median_house_value	,wowww it is our label
```

```python
#Adding a columns name -income_cat(income catagories) into df (by applying function in median_income)
df['income_cat'] = pd.cut(df['median_income'],bins=[0,1.5,3.0,4.5,6.0,np.inf],labels=[1,2,3,4,5])
# adding a income_cat column into df in this column it give 
# labels->1 which which has meadian_income under 0 to 1.5.
# labels->2 which which has meadian_income under 1.5 to 3.0.
# labels->3 which which has meadian_income under 3.0 to 4.5.
# labels->4 which which has meadian_income under 4.5 to 6.0.
# labels->5 which which has meadian_income 6.0 to above.
```

```python
df
```

```python
#anathor method--Stratifed shuffle split
```

```python
#write code this method
from sklearn.model_selection import StratifiedShuffleSplit
split = StratifiedShuffleSplit(n_splits=1,test_size=0.2,random_state=42)
for train_index,test_index in split.split(df,df['income_cat']):   #split amti karibu train set au test set deba kintu income_cat columns ta sabu thire asiba au same asiba
    strat_train_set = df.loc[train_index]
    strat_test_set = df.loc[test_index]
```

```python
strat_train_set
```

```python
strat_test_set
```

```python
#lets remove the income_cat column
for sett in(strat_train_set,strat_test_set):
    sett.drop("income_cat",axis=1,inplace=True)
```

```python
#main-part
#loading the train set into df 
df = strat_train_set   #ame athi jaha operation or taking feature or labels from the train data .don't tuching the test data 
```
